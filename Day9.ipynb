```python
def find_invalid( l, preamble_len ):
    for i in range(0, len( l ) - preamble_len):
        
        preamble = l[i:preamble_len + i]
        digit = l[preamble_len + i]
        
        valid = digit in {x + y for x in preamble for y in preamble if x < y}
        if not valid:
            return digit
    
    return None

find_invalid( test, 5 )
```

```python
def find_range( l, target ):
    for i in range(len(l)):
        for j in range(i + 1, len(l) - i):
            slyce = l[i:j]
            summ = sum(slyce)
            if summ == target:
                return slyce
    
    return None
            
find_range( test, 127 )
```
